# Reading Data in

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = 'https://raw.githubusercontent.com/kmath22/kmath22/main/Netflix.csv'

netflix_df = pd.read_csv(data)

netflix_df.head()
```

# Pie Charts

```python
netflix_df['Subscription Type'].value_counts().plot(kind = 'pie')
```

```python
#Add title
plt.title("Netflix Subscription Type")
netflix_df['Subscription Type'].value_counts().plot(kind = 'pie')

```

```python
#Change angles
plt.title("Netflix Subscription Type")
netflix_df['Subscription Type'].value_counts().plot(kind = 'pie', startangle = 90)
```

```python
#Add percents
plt.title("Netflix Subscription Type")
netflix_df['Subscription Type'].value_counts().plot(kind = 'pie', autopct='%1.1f%%')
```

```python
#Change colors
plt.title("Netflix Subscription Type")
netflix_df['Subscription Type'].value_counts().plot(kind = 'pie', colors=['olivedrab', 'rosybrown', 'saddlebrown'])
```

```python
#Explode feature with Legend
split = (0, 0.1, 0.2)
netflix_df['Subscription Type'].value_counts().plot(kind = 'pie', explode = split, shadow=True, startangle=90)
plt.legend(loc ="lower left")
```

# **Bar** **Chart**

```python
plt.title("Netflix Devices")
netflix_df['Device'].value_counts().plot(kind = 'bar')
```

```python
#Change color and width
netflix_df['Device'].value_counts().plot(kind = 'bar', color = 'gray', width = 0.2)
```

```python
#Add axis
plt.title("Netflix Devices")
netflix_df['Device'].value_counts(normalize = True).plot(kind = 'bar', color = 'orange', width = 0.5)
plt.xlabel("Types of Devices")
plt.ylabel("percentage")
```

```python
#Add gridlines and multiples color
plt.title("Netflix Devices")
netflix_df['Device'].value_counts(normalize = True).plot(kind = 'barh', color = ['green','gray','gray','red'], width = 0.9)
plt.ylabel("Types of Devices")
plt.xlabel("percentage")
plt.grid()
plt.legend()
```

# **Boxplot in the Seaborn Library**

```python
import seaborn as sns
```

```python
plt.title("Netflix Monthly Revenue")
sns.boxplot(x = 'Monthly Revenue' , data = netflix_df)
plt.grid(axis = 'x' ,linestyle = '--', linewidth = 0.4)
```

```python
sns.set(style="darkgrid")
```

```python
plt.title("Netflix Monthly Revenue")
sns.boxplot(x = 'Monthly Revenue' , data = netflix_df)
```

```python
plt.title("Netflix Monthly Revenue")
sns.boxplot(x = 'Age' , y = 'Device', data = netflix_df)
plt.ylabel("Type of Devices")
plt.xlabel("Ages")
```

```python
plt.title("Netflix Monthly Revenue")
sns.boxplot(x = 'Age' , y = 'Device', data = netflix_df,  palette="Reds")
plt.axvline(x = 37, color = 'r', linestyle = '--')
```
